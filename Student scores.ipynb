### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('drive/My Drive/Colab Notebooks/Sparks/Regression/student_scores.csv')
```

```python
df.head()
```

### Assign variables and split dataset into train and test

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Build model

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
print(f"intercept: {lr.intercept_:.4f}")
```

### make predictions on test dataset and evaluate

```python
# make predictions on test set
predictions_test = lr.predict(X_test)
```

```python
# get metrics - R^2, MAE, RMSE
r2_test = r2_score(y_test, predictions_test)
mae_test = mean_absolute_error(y_test, predictions_test)
rmse_test = mean_squared_error(y_test, predictions_test)
```

```python
# print metrics
print("metrics for test data")
print(f"R^2 score : {r2_test:.4f}")
print(f"mae       : {mae_test:.4f}")
print(f"rmse      : {rmse_test:.4f}")
```

### plot predictions against actual values

```python
plt.scatter(y_test, predictions_test)
plt.title("test dataset")
plt.xlabel("actual")
plt.ylabel("predicted")
plt.gca().axline([7, 7], [11, 11], color="red")
plt.grid()

plt.show()
```
